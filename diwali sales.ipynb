```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sas
```

```python
df = pd.read_excel("diwali.xlsx")
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df.isna().sum()
```

```python
###this is the analysis of male and female in sort comaparison
```

```python
ax = sas.countplot(x='Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#now the analysis is for like which gender spent more monney 
df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sas.barplot(x = 'Gender',y = 'Amount', data = sales_gen)
```

```python
#compariason of age grp on basis of male and female
ax = sas.countplot(data = df,x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total number of states and top 10 states 
my_color = ['orange','blue','red','green','black']
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
sas.set(rc={'figure.figsize':(20,10)})
sas.barplot(x = 'State',y = 'Orders', data = sales_state, palette = my_color)
```

```python
my_color = ['orange','blue','red','green','black']
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sas.set(rc={'figure.figsize':(20,10)})
sas.barplot(x = 'State',y = 'Amount', data = sales_state, palette = my_color)
```

```python
# here we see the maritial status annd from thst find who buy more product
ax = sas.countplot(data = df,x = 'Marital_Status')

sas.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sas.set(rc={'figure.figsize':(6,5)})
sas.barplot(x = 'Marital_Status',y = 'Amount', data = sales_state,hue = 'Gender')
```

```python
#here we are finding that which ocupaton people by most prodct
ax = sas.countplot(data = df,x = 'Occupation')

sas.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sas.countplot(data = df,x = 'Product_Category')

sas.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
